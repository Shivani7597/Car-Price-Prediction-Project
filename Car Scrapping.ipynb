### 1st

```python
#Activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\MUNGI\Downloads\chromedriver_win32 (1)\chromedriver.exe")

#Opening olx web page for data extraction
driver.get('https://www.olx.in/cars_c84')
time.sleep(4)

#loading pages until there is no option for "load more" available on the webpage
while True:
    try:
        loadMoreButton = driver.find_element_by_xpath("//div[@class='JbJAl']/button")
        time.sleep(2)
        loadMoreButton.click()
        time.sleep(5)
    except Exception as e:
        print('e')
        break
print("Complete")
time.sleep(5)
```

```python
#Fetching urls
urls=[]
for i in driver.find_elements_by_xpath("//ul[@class='rl3f9 _3mXOU']/li/a"):
    urls.append(i.get_attribute("href"))
```

```python
#Urls scrapped
urls
```

```python
#length of url scrapped
len(urls)
```

```python
#Creating set of empty lists
brand=[]
model=[]
variant=[]
year=[]
fuel=[]
transmission=[]
km=[]
owners=[]
location=[]
price=[]
```

```python
#Fetching data from the urls
for i in urls:
    driver.get(i)
    time.sleep(5)
    
        
    #Fetching Brand name
    try:
        title_chk = driver.find_element_by_xpath('//h3[@data-aut-id="itemDescriptonTitle"][1]/span')
        if title_chk.text != "Details" : raise NoSuchElementException
        bat_chk = driver.find_element_by_xpath('//span[@data-aut-id="key_make"]')
        if bat_chk.text != "Brand" : raise NoSuchElementException
        mn=driver.find_element_by_xpath('//span[@data-aut-id="value_make"]')
        if title_chk.text == "Details":
            if bat_chk.text == "Brand":
                brand.append(mn.text.replace("\n","-"))
    except NoSuchElementException as e:
        brand.append('-')
        
    #Fetching Model Name
    try:
        title_chk1 = driver.find_element_by_xpath('//h3[@data-aut-id="itemDescriptonTitle"][1]/span')
        if title_chk1.text != "Details" : raise NoSuchElementException
        bat_chk1 = driver.find_element_by_xpath('//span[@data-aut-id="key_model"]')
        if bat_chk1.text != "Model" : raise NoSuchElementException
        o=driver.find_element_by_xpath('//span[@data-aut-id="value_model"]')
        if title_chk1.text == "Details":
            if bat_chk1.text == "Model":
                model.append(o.text.replace("\n","-"))
    except NoSuchElementException as e:
        model.append('-')
        
        #Fetching Variant Name
    try:
        title_chk2 = driver.find_element_by_xpath('//h3[@data-aut-id="itemDescriptonTitle"][1]/span')
        if title_chk2.text != "Details" : raise NoSuchElementException
        bat_chk2 = driver.find_element_by_xpath('//span[@data-aut-id="key_variant"]')
        if bat_chk2.text != "Variant" : raise NoSuchElementException
        o=driver.find_element_by_xpath('//span[@data-aut-id="value_variant"]')
        if title_chk2.text == "Details":
            if bat_chk2.text == "Variant":
                variant.append(o.text.replace("\n","-"))
    except NoSuchElementException as e:
        variant.append('-')
        
     #Fetching Manufacturing Year
    try:
        title_chk3 = driver.find_element_by_xpath('//h3[@data-aut-id="itemDescriptonTitle"][1]/span')
        if title_chk3.text != "Details" : raise NoSuchElementException
        bat_chk3 = driver.find_element_by_xpath('//span[@data-aut-id="key_year"]')
        if bat_chk3.text != "Year" : raise NoSuchElementException
        dr=driver.find_element_by_xpath('//span[@data-aut-id="value_year"]')
        if title_chk3.text == "Details":
            if bat_chk3.text == "Year":
                year.append(dr.text)
    except NoSuchElementException as e:
        year.append('-')
        
    #Fetching Fuel
    try:
        title_chk4 = driver.find_element_by_xpath('//h3[@data-aut-id="itemDescriptonTitle"][1]/span')
        if title_chk4.text != "Details" : raise NoSuchElementException
        bat_chk4 = driver.find_element_by_xpath('//span[@data-aut-id="key_petrol"]')
        if bat_chk4.text != "Fuel" : raise NoSuchElementException
        ds=driver.find_element_by_xpath('//span[@data-aut-id="value_petrol"]')
        if title_chk4.text == "Details":
            if bat_chk4.text == "Fuel":
                fuel.append(ds.text)
    except NoSuchElementException as e:
        fuel.append('-')
        
    #Fetching Transmission
    try:
        title_chk5 = driver.find_element_by_xpath('//h3[@data-aut-id="itemDescriptonTitle"][1]/span')
        if title_chk5.text != "Details" : raise NoSuchElementException
        bat_chk5 = driver.find_element_by_xpath('//span[@data-aut-id="key_transmission"]')
        if bat_chk5.text != "Transmission" : raise NoSuchElementException
        r=driver.find_element_by_xpath('//span[@data-aut-id="value_transmission"]')
        if title_chk5.text == "Details":
            if bat_chk5.text == "Transmission":
                transmission.append(r.text)
    except NoSuchElementException as e:
        transmission.append('-')
        
    #Fetching KM Driven 
    try:
        title_chk6 = driver.find_element_by_xpath('//h3[@data-aut-id="itemDescriptonTitle"][1]/span')
        if title_chk6.text != "Details" : raise NoSuchElementException
        bat_chk6 = driver.find_element_by_xpath('//span[@data-aut-id="key_mileage"]')
        if bat_chk6.text != "KM driven" : raise NoSuchElementException
        rt=driver.find_element_by_xpath("//span[@data-aut-id='value_mileage']")
        if title_chk6.text == "Details":
            if bat_chk6.text == "KM driven":
                km.append(rt.text)
    except NoSuchElementException as e:
        km.append('-')
                                             
    #Fetching No. of Owners
    try:
        title_chk7 = driver.find_element_by_xpath('//h3[@data-aut-id="itemDescriptonTitle"][1]/span')
        if title_chk7.text != "Details" : raise NoSuchElementException
        bat_chk7 = driver.find_element_by_xpath('//span[@data-aut-id="key_first_owner"]')
        if bat_chk7.text != "No. of Owners" : raise NoSuchElementException
        lw=driver.find_element_by_xpath('//span[@data-aut-id="value_first_owner"]')
        if title_chk7.text == "Details":
            if bat_chk7.text == "No. of Owners":
                owners.append(lw.text.replace("\n","-"))
    except NoSuchElementException as e:
        owners.append('-')
        
    #Fetching Location
    try:
        ln=driver.find_element_by_xpath("//div[@data-aut-id='itemLocation']/div/span")
        location.append(ln.text.replace("\n"," "))
    except NoSuchElementException as e:
        location.append('-')
        
    #Fetching Price
    try:
        pr=driver.find_element_by_xpath("//section[@class='_2wMiF']/span[1]")
        price.append(pr.text.replace("\n"," "))
    except NoSuchElementException as e:
        price.append('-')
        
        
   
```
